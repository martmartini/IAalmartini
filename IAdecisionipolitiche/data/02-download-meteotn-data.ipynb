```python
import pandas as pd
```

```python
import io
```

```python
columns = [
    "TIME",
    "VAL",
    "QUAL",
    "COMNT",
    "SITE",
    "VAR",
    "DT",
    "BLOCK",
    "CRDATE",
    "CRPROG",
    "CRUSER",
    "MDDATE",
    "MDPROG",
    "MDUSER",
    "ARDATE",
    "ARPROG",
    "ARUSER",
    "WRDATE",
    "WRPROG",
    "WRUSER",
]
```

```python
buf = io.StringIO("\n".join(lines[:-1]))
df = pd.read_csv(buf, sep=",", header=None, names=columns, parse_dates=["TIME"])
df
```

```python
df.VAR.unique
```

```python
df_p = pd.read_csv(
    "../data/mt/raw/t0408_p.csv",
    sep=",",
    skiprows=4,
    header=None,
    names=["Data", "Valore", "Qualità", "Info"],
    index_col=None,
).drop(columns=["Info"])

df_p["Variabile"] = "Precipitazione"
df_p["Unità di misura"] = "mm"
df_p["Data"] = pd.to_datetime(df_p.Data, format="%H:%M:%S %d/%m/%Y")

df_p
```

```python
df_pa = pd.read_csv(
    "../data/mt/raw/t0408_pa.csv",
    sep=",",
    skiprows=4,
    header=None,
    names=["Data", "Valore", "Qualità", "Info"],
    index_col=None,
).drop(columns=["Info"])

df_pa["Variabile"] = "Pressione Atmosferica"
df_pa["Unità di misura"] = "hPa"
df_pa["Data"] = pd.to_datetime(df_pa.Data, format="%H:%M:%S %d/%m/%Y")

df_pa
```

```python
df_ta = pd.read_csv(
    "../data/mt/raw/t0408_ta.csv",
    sep=",",
    skiprows=4,
    header=None,
    names=["Data", "Valore", "Qualità", "Info"],
    index_col=None,
    encoding="latin1",
).drop(columns=["Info"])

df_ta["Variabile"] = "Temperatura"
df_ta["Unità di misura"] = "°C"
df_ta["Data"] = pd.to_datetime(df_ta.Data, format="%H:%M:%S %d/%m/%Y")

df_ta
```

```python
df = pd.concat([df_p, df_pa, df_ta])
df
```
